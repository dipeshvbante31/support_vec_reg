## feature encodeing(label Encoding and OneHotEncoding)

```python
###independent and dependent feature 

df.columns
```

```python
x=df[['total_bill', 'tip', 'sex', 'smoker', 'day', 'time', 'size']]
y=df['total_bill']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
x_train,y_train
```

```python
### feature encoding 

from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
x_train['sex']=le1.fit_transform(x_train['sex'])
x_train['smoker']=le2.fit_transform(x_train['smoker'])
x_train['time']=le3.fit_transform(x_train['time'])
```

```python
x_train.head()
```

```python
x_test['sex']=le1.transform(x_test['sex'])
x_test['smoker']=le2.transform(x_test['smoker'])
x_test['time']=le3.transform(x_test['time'])
```

```python
x_test.head()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[4])],remainder='passthrough')
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
x_train=ct.fit_transform(x_train)
x_test=ct.transform(x_test)
```

## support vector regression

```python
from sklearn.svm import SVR
svr=SVR()
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred=svr.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
param_grid={
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
}
```

```python
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid_predict=grid.predict(x_test)
print(r2_score(y_test,grid_predict))
print(mean_absolute_error(y_test,grid_predict))
print(mean_squared_error(y_test,grid_predict))
```
